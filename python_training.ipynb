```python
# Creating a set
my_set = {1, 2, 3, 4, 5}
print(f"Initial Set: {my_set}")

# Sets automatically handle unique elements (duplicates are ignored)
set_with_duplicates = {1, 2, 2, 3, 4, 4, 5}
print(f"Set with duplicates provided: {set_with_duplicates}")

# Adding elements to a set
my_set.add(6)
my_set.add(1) # Adding an existing element has no effect
print(f"Set after adding elements: {my_set}")

# Removing elements from a set
my_set.remove(3) # Removes the element 3
# my_set.remove(10) # This would raise a KeyError if 10 is not in the set
print(f"Set after removing an element: {my_set}")

my_set.discard(2) # Removes 2 if it exists, does nothing if it doesn't exist
my_set.discard(10) # Does not raise an error
print(f"Set after discarding an element: {my_set}")

# Pop an arbitrary element (sets are unordered)
popped_element = my_set.pop()
print(f"Popped element: {popped_element}")
print(f"Set after popping an element: {my_set}")
```

```python

# Basic set operations
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

# Union: elements in either set_a or set_b or both
union_set = set_a.union(set_b)  # or set_a | set_b
print(f"Union of set_a and set_b: {union_set}")

# Intersection: elements common to both set_a and set_b
intersection_set = set_a.intersection(set_b) # or set_a & set_b
print(f"Intersection of set_a and set_b: {intersection_set}")

# Difference: elements in set_a but not in set_b
difference_set = set_a.difference(set_b) # or set_a - set_b
print(f"Difference (set_a - set_b): {difference_set}")

# Symmetric difference: elements in either set but not in both
symmetric_difference_set = set_a.symmetric_difference(set_b) # or set_a ^ set_b
print(f"Symmetric Difference: {symmetric_difference_set}")
```

```python
# Basic set operations
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
# Check for subsets and supersets
set_c = {1, 2}
is_subset = set_c.issubset(set_a)
is_superset = set_a.issuperset(set_c)
print(f"Is set_c a subset of set_a? {is_subset}")
print(f"Is set_a a superset of set_c? {is_superset}")
```

```python
# Creating a dictionary
my_dict = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}
print(f"Initial Dictionary: {my_dict}")

# Accessing values by key
print(f"Name: {my_dict['name']}")
print(f"Age: {my_dict.get('age')}") # Using .get() is safer as it returns None for non-existent keys

# Modifying a value
my_dict['age'] = 31
print(f"Dictionary after modifying age: {my_dict}")

# Adding a new key-value pair
my_dict['occupation'] = "Engineer"
print(f"Dictionary after adding occupation: {my_dict}")

# Removing a key-value pair
del my_dict['city']
print(f"Dictionary after deleting city: {my_dict}")

# Pop a key-value pair (removes and returns the value)
popped_occupation = my_dict.pop('occupation')
print(f"Popped occupation: {popped_occupation}")
print(f"Dictionary after popping occupation: {my_dict}")

# Get all keys
print(f"Keys: {my_dict.keys()}")

# Get all values
print(f"Values: {my_dict.values()}")

# Get all items (key-value pairs)
print(f"Items: {my_dict.items()}")

# Iterating through a dictionary
print("\nIterating through dictionary:")
for key, value in my_dict.items():
    print(f"{key}: {value}")

# Check if a key exists
print(f"Is 'name' in dictionary? {'name' in my_dict}")
print(f"Is 'city' in dictionary? {'city' in my_dict}")

# Clearing a dictionary
my_dict.clear()
print(f"Dictionary after clearing: {my_dict}")
```

```python
# Defining a simple function without parameters
def greet():
    print("Hello from the function!")

# Calling the function
greet()

# Defining a function with parameters
def add_numbers(a, b):
    result = a + b
    print(f"The sum of {a} and {b} is {result}")

# Calling the function with arguments
add_numbers(5, 3)
add_numbers(10, 20)

# Defining a function that returns a value
def multiply(x, y):
    return x * y

# Calling the function and storing the result
product = multiply(4, 6)
print(f"The product is: {product}")

# Function with default parameter value
def say_hello(name="Guest"):
    print(f"Hello, {name}!")

say_hello()        # Uses default value
say_hello("Alice") # Overrides default value

# Function with variable number of arguments (*args)
def sum_all(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(f"Sum of 1, 2, 3: {sum_all(1, 2, 3)}")
print(f"Sum of 10, 20, 30, 40: {sum_all(10, 20, 30, 40)}")
```

```python
#factorial usnig recursion
def fact(n):
  # Base Case: If n is 0, the factorial is 1. This stops the recursion.
  if n==0:
    return 1
  # Recursive Step: If n is not 0, it returns n multiplied by the factorial of (n-1).
  # The function calls itself with a smaller argument until it reaches the base case.
  return n*fact(n-1)

# Calling the function to calculate the factorial of 3
print(fact(3))
```

```python

#parent class
class Animal:
    def speak(self):
        print("Animal makes a sound")

#child class
class Dog(Animal):
    def speak(self):
        print("Dog barks")
#child class
class Cat(Animal):
    def speak(self):
        print("Cat meows")

# Create objects
animal = Animal()
dog = Dog()
cat = Cat()

animal.speak()
dog.speak()
cat.speak()
```

```python
#multiple inheritance
class A:
  def fun1(self):
    print("fun1")
class B:
  def fun2(self):
    print("fun2")
class C(A,B):
  def fun3(self):
    print("fun3")
obj=C()
obj.fun1()
obj.fun2()
obj.fun3()


```

```python
#one parent -> multi children
class A:
  def show_A(self):
    print("A")
class B(A):
  def show_B(self):
    print("B")
class C(A):
  def show_C(self):
    print("C")
  def show_B (A):
    print("B")
class show_C (A):
  pass
x=C()
x.show_A()
x.show_B()
x.show_C()
```

```python
#implement of parent class method
class Father:
  def show(self):
    print("this is Father class")
class Son(Father):
  def show(self):
    print("this is Son class")
    super().show()     #super() ke use se same class access kar skate hai
x=Son()
x.show()
```
